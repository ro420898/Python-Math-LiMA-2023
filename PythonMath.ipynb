Tarea 01 : Hallar el exponente n más pequeño tal que el módulo de (3+4i)^n es mayor que 100.
=================

**Date:** 2023-08-14



```python
z = 3+4j
m = 100

def find_n(z, m): #Regresa el exponente n más pequeño tal que el módulo de z**n es mayor que m.
  for n in range(1, m + 1):
    if abs(z ** n) > m:
      return n
  raise ValueError(f'No hay ningún exponente n tal que el módulo de {z}^n sea mayor que {m}.')

n = find_n(z, m)

print(f'El exponente n más pequeño tal que el módulo de {z}^n es mayor que {m} es {n}.')
print('z =', z,'\nn =', n,'\nz^n =', z**n,'\n|z| =', abs(z**n))
```
